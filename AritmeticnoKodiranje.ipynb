# Aritmetično kodiranje in dekodiranje

```python
def aritmeticnoKodiranje(verjetnosti, zaporedje, interval, reverse=False):
    if reverse:  # če želimo, da je stop znak na vrhu
        verjetnosti.reverse()

    output = ""
    for idx, j in enumerate(zaporedje):
        razpon = (interval[1] - interval[0])
        new_int = []
        next_b = interval[0]

        output += f"\nKorak {idx+1}, člen: {j}, začetni interval: [{interval[0]:.5f}, {interval[1]:.5f}]\n"
        output += "  Simbol |     Podinterval\n"
        output += "  -------+--------------------\n"

        for i, verjetnost in enumerate(verjetnosti):
            dolzina = razpon * verjetnost
            temp = [next_b, next_b + dolzina]
            new_int.append(temp)
            oznaka = " <== izbran" if (not reverse and i == j-1) or (reverse and i == len(verjetnosti) - j) else ""
            output += f"    {i+1:2d}    |  [{temp[0]:.5f}, {temp[1]:.5f}]{oznaka}\n"
            next_b = temp[1]

        if not reverse:
            interval = new_int[j-1]
        else:
            interval = new_int[len(verjetnosti) - j]

    koncna_stevilka = (interval[0] + interval[1]) / 2
    output += f"\nKodirano število (sredina zadnjega intervala): {koncna_stevilka:.10f}\n"

    print(output)
    return koncna_stevilka
verjetnosti = [4/9,1/9,3/9,1/9]
zaporedje = [2,3,3,1,4]
interval = [0.0,1.0]
reverse = False
koda = aritmeticnoKodiranje(verjetnosti,zaporedje,interval,reverse)
```

```python
def aritmeticnoDekodiranje(verjetnosti, kodirano_stevilo, interval, reverse=False):
    verjetnosti = list(reversed(verjetnosti)) if reverse else verjetnosti
    stop_indeks = 0 if reverse else len(verjetnosti) - 1

    output = ""
    zaporedje = []
    korak = 1

    while True:
        razpon = interval[1] - interval[0]
        new_int = []
        next_b = interval[0]

        output += f"\nKorak {korak}, vrednost: {kodirano_stevilo:.10f}, interval: [{interval[0]:.10f}, {interval[1]:.10f}]\n"
        output += "  Simbol |     Podinterval\n"
        output += "  -------+--------------------\n"

        izbran_indeks = None
        for i, verjetnost in enumerate(verjetnosti):
            dolzina_int = razpon * verjetnost
            temp = [next_b, next_b + dolzina_int]
            new_int.append(temp)

            if temp[0] <= kodirano_stevilo < temp[1]:
                izbran_indeks = i
                oznaka = " <== izbran"
            else:
                oznaka = ""

            output += f"    {i+1:2d}    |  [{temp[0]:.10f}, {temp[1]:.10f}]{oznaka}\n"
            next_b = temp[1]

        if izbran_indeks is None:
            output += "\nNapaka: kodirano število ne ustreza nobenemu podintervalu.\n"
            break

        simbol_index = izbran_indeks if not reverse else len(verjetnosti) - izbran_indeks - 1
        zaporedje.append(simbol_index + 1)
        interval = new_int[izbran_indeks]

        if izbran_indeks == stop_indeks:
            output += f"\nPrepoznan stop znak (simbol {simbol_index + 1}) – dekodiranje zaključeno.\n"
            break

        korak += 1

    print(output)
    return zaporedje

aritmeticnoDekodiranje(verjetnosti, koda, interval)
```
